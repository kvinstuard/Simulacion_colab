#1. Implementar los algoritmos de generación de secuencias pseudoaletorias (Von Neuman, congruencia lineal y fibonacci), ejecutarlos y obtener sus salidas en alguna(s) variable(s).

Generar secuencias de 1000 datos.



##Von Neumann

```python
def von_neumann(semilla, i):
    resultado = []
    for i in range(i):
        cuadrado = semilla ** 2
        # convirtiendo la semilla a string y rellenando con ceros 
        cuadrado_str = str(cuadrado).zfill(8)
        #Extrayendo los 4 digitos de la mitad
        mitad = cuadrado_str[2:6]
        #Convirtiendo los digitos a entero
        extraido = int(mitad)
        #Haciendo los digitos extraidos la nueva semilla
        semilla = extraido
        #Normalizando
        normalizar = extraido / 10000
        # Añadiendo los numeros a la lista
        resultado.append(normalizar)
    return resultado
```

El la semilla 1009 hace que se degenere a cero rapidamente.

```python
resultToZero = von_neumann(1009, 1000)
resultRamdon = von_neumann(3708,1000)
print(resultToZero)
print(resultRamdon)
```

##Congruencia lineal

```python
def congruencia(semilla, a, c, m, iteraciones):
    resultado = []
    for i in range(iteraciones):
        semilla = (a * semilla + c) % m
        # Normalizando
        normalizado = semilla / m
        # Añadir al array
        resultado.append(normalizado)
    return resultado
```

```python
import matplotlib.pyplot as plt

# Datos
result = congruencia(5, 106, 1283, 6075, 10000)

plt.hist(result, bins=100, edgecolor='black')

plt.title('Histograma')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

plt.show()
```

##Fibonacci

```python
def fibonacci(n, j, k, op, semilla, m):
    # Inicializamos los valores
    fib = [semilla, semilla + 1]

    for i in range(2, n):
        xn = 0
        if op == "+":
            xn = fib[i-j] + fib[i-k]
        elif op == "*":
            xn = fib[i-j] * fib[i-k]
        xn %= m
        fib.append(xn)
    return fib

# Ejemplo:
secuencia = fibonacci(1000, 2, 3, "+", 1, 100)
print(secuencia)
```

```python
import matplotlib.pyplot as plt

# Graficar de la secuencia
plt.plot(secuencia)
plt.xlabel('Índice')
plt.ylabel('Número pseudoaleatorio')
plt.title('Secuencia pseudoaleatoria generada por Fibonacci')
plt.show()
```

#2. Implementar los algoritmos de pruebas de uniformidad ( Kolmogorov-Smirnov, χ2)

##Chi_cuadrado: realiza prueba de uniformidad dado un nivel de significancia.

    Parametros:
    FO (array): Array de frecuencias observadas.
    FE (array): Array de frecuencias esperadas.
    alpha (float): Nivel de confianza.
    
    Retorna:
    p-value (float): El valor de p probabilidad.
    reject (bool):True si la hipotesis nula es rechazada, False de lo contrario.

```python
import numpy as np
from scipy.stats import chi2

def chi_cuadrado(secuencia, k, alpha=0.05):
    n = len(secuencia)
    FO, _ = np.histogram(secuencia, k)
    FE = np.ones(k) * (n / k)
    chi_sq_stat = np.sum((FO - FE)**2 / FE)
    df = k - 1 
    p_value = 1 - chi2.cdf(chi_sq_stat, df)
    reject = p_value < alpha
    return p_value, reject

```

Ejemplo donde **NO** se rechaza la hipotesis.

```python
observado = np.array([11,6,16,9,7,11,10,9,15,6])
p_value, reject = chi_cuadrado(observado, 10, alpha=0.05)
print("p-value:", p_value)
print("Rechazar hipotesis nula?", reject)
```

##Kolmogorov_Smirnov: realiza prueba de uniformidad dado un nivel de significancia.

    Parametros:
    lista(array): Array con numeros pseudoaleatorios
    alpha (float): Nivel de confianza.
    
    Retorna:
    Un string segun la condicion.

```python
from scipy.stats import kstest
import numpy as np

def kolmogorov_Smirnov(lista, alpha):
    n = len(lista)
    cdf = lambda x: x # Definimos la función de distribución acumulada esperada
    ks_stat, p_value = kstest(lista, cdf) 
    critical_value = 1.36 / np.sqrt(n) # Calculamos el valor crítico para el nivel de confianza 0.05
    if ks_stat > critical_value:
        return "Rechazar hipótesis nula. Los datos no siguen una distribución uniforme."
    else:
        return "No se puede rechazar hipótesis nula. Los datos siguen una distribución uniforme."
```

```python
datos = [0.11,0.17,0.33,0.42,0.49,0.60,0.70,0.79,0.94,1]
resultado = kolmogorov_Smirnov(datos, 0.05)
print(resultado)
```

#2.Implementar los algoritmos de pruebas de independecia (corridas, series y poker)

##Corridas: realiza prueba de independecia dado un nivel de significancia.

    Parametros:
    data(array): Array con datos observados de la variable aleatoria
    alpha (float): Nivel de confianza.
    
    Retorna:
    Un string segun la condicion.



```python
def corridas(data, alpha):
    n = len(data) #numero observaciones
    m = np.mean(data) #media de los datos
    s = np.std(data)#desviacion estandar de los datos
    b = 0 #contador
    for i in range(1, n):
        if (data[i] - m) * (data[i-1] - m) < 0:
            b += 1
    z = (b - (2 * n - 1) / 3) / np.sqrt(16 * n - 29)
    critical_value = 1.96 
    if abs(z) > critical_value:
        return "Rechazar hipótesis nula. Los datos no son independientes."
    else:
        return "No se puede rechazar hipótesis nula. Los datos son independientes."
```

 ## Poker: Realiza prueba de independecia tomando en cuenta un nivel de confianza

    Parametros:
    secuencia(array): Array con secuencia de datos pseudoaleatorios
    
    Retorna:
    Un booleano.

```python
import numpy as np
from scipy.stats import chi2

def prueba_independencia_poker(secuencia):
    n = len(secuencia)
    tipos = [0] * 32
    for i in range(0, n, 5):
        bloque = secuencia[i:i+5]
        if len(set(bloque)) == 1:
            tipos[0] += 1
        elif len(set(bloque)) == 2:
            tipos[1] += 1
        elif len(set(bloque)) == 3:
            if bloque.count(bloque[0]) == 3:
                tipos[2] += 1
            else:
                tipos[3] += 1
        elif len(set(bloque)) == 4:
            tipos[4] += 1
        else:
            tipos[5] += 1
    pi = [0.3240, 0.5040, 0.1080, 0.0720, 0.0090, 0.0045]
    E = [n/5 * pi[i] for i in range(6)]
    O = tipos
    X2 = sum((O[i] - E[i])**2 / E[i] for i in range(6))
    X2_alpha = chi2.ppf(1 - 0.05, 31)
    if (X2 < X2_alpha):
      return print("La secuencia pasa la prueba de independencia poker")
    else:
      return print("La secuencia no pasa la prueba de independencia poker")
```
